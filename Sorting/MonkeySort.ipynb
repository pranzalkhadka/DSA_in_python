```python
def is_sorted(array):

    sorted = True

    for i in range(len(array)-1):
        if array[i] > array[i+1]:
            sorted = False

    return sorted
```

```python
arr = [1,2,3,4,9,6]
```

```python
is_sorted(arr)
```

Monkey sort is a technique where if a monkey is given a set of cards to sort, it will shuffle randomly until the cards are sorted. So the worst case time complexity for this algo is infinite since random shuffle could never lead to sorted array.

```python
import random
```

```python
arr = [2,4,1,5,8,6]
```

```python
def monkey_sort(arr):

    while not is_sorted(arr):
        random.shuffle(arr)
        print(arr)

    print(arr)
```

```python
monkey_sort(arr)
```
